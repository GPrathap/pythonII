With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it.

Colab සමඟින් ඔබට දත්ත විශ්ලේෂණය කිරීමට සහ දෘශ්‍යමාන කිරීමට ජනප්‍රිය Python Library සම්පූර්ණ බලය භාවිතා කළ හැක. පහත කේත කොටුව අහඹු දත්ත ජනනය කිරීමට **numpy** භාවිත කරන අතර එය දෘශ්‍යමාන කිරීමට **matplotlib** භාවිත කරයි.

```
import numpy as np
from matplotlib import pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
```

Let's create a simple Python function, and call it from a loop

අපි සරල Python ශ්function ක් සාදා එය ලූපයකින් අමතන්න

```
def HelloWorldXY(x, y):
  if (x*y < 10):
    print("x*y was < 10")
  elif (x*y < 20):
    print("x*y was >= 10 but < 20")
  else:
    print("x*y was >= 20")
  return x*y

for i in range(1, 50, 10):  # (start, stop, step)
  r = HelloWorldXY(i,i)
  print("result x*y: {}".format(r))
```
