## **1.1 What Is Line Plot?**

## **Line Plot එකක් යනු කුමක්ද?**

> **Line plot is a type of chart that displays information as a series of data points connected by straight line segments**

Line plots are generally used to visualize the directional movement of one or more data over time. In this case, the X axis would be datetime and the y axis contains the measured quantity, like, stock price, weather, monthly sales, etc.

> **A line plot is often the first plot of choice to visualize any time series data**.

> **Line Plot එකක් යනු සරල රේඛා ඛණ්ඩ මගින් සම්බන්ධ වූ ලක්ෂ සමූහයක විචලනය පෙන්වන ප්‍රස්ථාර වර්ගයක්.**

අපි  සාමාන්යෙන් කාලයත් සමග දත්ත වල දිශානුගත චලනය පෙන්වීමට Line  Plots භාවිතා කරනවා . මෙම අවස්ථාවන්හීදී , X අක්ෂය කාලය සම්බන්ද රාශියක්  වන අතර y අක්ෂය මනින ලද රාශියක් වෙන්න පුලුවන්. උදාහරණ විදිහට කොටසක විකුණුම්  මිල, යම් පුද්ගයකුගේ සාමාන්‍ය වැටුප, මාසික විකුණුම් ප්‍රමාණය  වැනි රාශීන් තියෙන්න පුලුවන්.

> **Line Plots කියන්නේ ගොඩක් වෙලාවට time series දත්ත පෙන්වීමට තෝරා ගන්නා පළමු තේරීම කියලා කියන්න පුලුවන්.**.

## **1.2 Simple Line Plots**
## **සරල Line Plot එකක් ඇඳීම**


As we all know, we need set of coordinates(X and Y coordinates) for creating a Line graph. Let's see by taking simple examples how to create a graph using Python.

ඔයාලා හැමෝම දැනටමත් දන්නවා , රේඛා ප්‍රස්ථාරයක් ඒ කියන්නේ Line Graph එකක් නිර්මාණය කිරීම සඳහා අපට ඛණ්ඩාංක (X සහ Y ඛණ්ඩාංක) අවශ්‍ය වෙනවා. ඒ කියන්නේ අපි දන්නවා අපිට Line graph එකක් අදින්න x ඛණ්ඩාංක සහ y ඛණ්ඩාංක අවශ්‍ය වෙනවා. දැන් අපි සරල උදාහරණ අරගෙන බලමු Phyton භාවිතා කරලා කොහොමද මේ Line graph එකක් නිර්මාණය කරගන්නේ කියලා.


### 1.2.1 **Example 1**: Simple Graph For Y=log(X)

### **උදාහරණ 1**: y = log (x) ඇදීම 


Let’s create a dataset with 50 values between 1 and 100 using the **np.linspace()** function. This will go in the X axis, whereas the Y axis values is the log of x.

The line graph of **y** vs **x** is created using **plt.plot(x,y)**. It joins all the points in a sequential order.

හරි අපි දැන් ඇඳල බලමු simple graph එකක් y = log x කියන ප්‍රස්ථාරයට. ඉස්සෙල්ලාම අපි බලමු ඛණ්ඩාංක ටික generate කරගන්නෙ කොහොමද කියලා. අපි x අක්ෂයට 1 ඉඳලා  100 දක්වා තියෙන values 50ක් np.linespace() function එකෙන් generate කරගමු. ඒ x අගයන් 50 use කරලා,ඒ x අගයට සමානුපාත  log (x), ඒ කියන්නේ y අගය හොයාගමු. ඔන්න ඊට පස්සේ අපිට දැන් x අගයනුත්, y අගයනුත්, ඔක්කොම ලෑස්තියි.ඊට පස්සේ අපිට පුළුවන් plt.plot(x,y)  function එක භාවිතා කරලා මේ දත්ත සමූහය රේඛා ප්‍රස්තාරයක ඇඳගන්න.


#### **Example 1**: Answer

#### උදාහරණ 1 : විසඳුම


````python
import numpy as np
from matplotlib import pyplot as plt
# Simple Line Plot

# change the style of graph,
# print(plt.style.available) #-> uncomment and see, various graph styles support.
plt.style.use('fivethirtyeight')


# first we have to get the coordinates of X and Y
x=np.linspace(1,100,50)

# Simple line plot with lesser points
# x=np.linspace(1,100,5)


y=np.log(x)

# Then simply plot X and Y coordinates
plt.plot(x,y)

# finally the graph can be named meaningfully.
plt.xlabel('x')
plt.ylabel('logx')
plt.title('Simple Line Plot')

# you can use xlim,ylim to see the graph within the specific range,try uncommenting 
plt.xlim(0,100)
plt.ylim(0,5)
````

Sometimes the the data you want to plot, may not be on the same scale.

For example, you want to plot the number of sales of a product and the number of enquires. Here, sales data could be in 10’s whereas the number of sales enquiries could be in 100’s. Still you want to compare their tandem movement (If one thing happens or done with another thing) by plotting both these lines in same chart.

It’s possible to plot for 2 different y axis(left and right) with different scales.

You can do this by using the twinx() function in python.

That is, after plotting the first chart using ax1, you can get the reference for the second Y axis (ax2) using the twinx() method of ax1.

මේ වගේ වෙලාවට අපිට පුළුවන් එකම x අක්ෂයට,  y අක්ෂ දෙකක් පාවිච්චි කරලා (සාමාන්‍ය විදිහට එකක් වම් පැත්තේ,අනිත් එක දකුණු පැත්තෙනුත්) ප්‍රස්තාර ඇඳගන්න.

මේ සඳහා  twinx() function එක පාවිච්චි කරන්න පුළුවන්.

පළමුව පහත දැක්වෙන විදිහට පළමු රේඛාව ax1 කියන reference එක use කරල ඇඳගෙන , දෙවන ප්‍රස්ථාරය ඇඳීමට y අක්ෂය සඳහා reference එක  (ax2), twinx() function එක use කරල ගන්න පුළුවන්.

#### **Example 1**: Answer
#### උදාහරණ 1 : විසඳුම

````python
#configure the graph size & style (height and width)

# Plot second line in secondary Y axis.
x = np.arange(1, 10, 0.1)
y1 = np.exp(x)
y2 = np.log(x)

# Plot
fig, ax1 = plt.subplots()
ax1.set_ylabel('exp')
ax1.plot(x, y1, color='blue')
ax1.set_ylim(0,40000)
ax1.grid()

# Second Line on secondary Y axis
ax2 = ax1.twinx()  # join a second axis with the first graph we just made
ax2.set_ylabel('log')
ax2.set_ylim(0,3)
ax2.plot(x, y2,color='red')
plt.show()
````

## **6.1  What is a Scatter plot?**
### **Scatter Plot එකක් යනු කුමක්ද?**

අපි දැන් විස්තරාත්මකව බලමු scatter plot එකක් කියන්නේ මොකක්ද කියලා

Scatter plot is a graph of two sets of data along the two axes. It is used to visualize the relationship between the two variables.

If the value along the Y axis seem to increase as X axis increases(or decreases), it could indicate a positive (or negative) linear relationship. Whereas, if the points are randomly distributed with no obvious pattern, it could possibly indicate a lack of dependent relationship.

In python matplotlib, the scatterplot can be created using the pyplot.plot() or the pyplot.scatter(). Using these functions, you can add more feature to your scatter plot, like changing the size, color or shape of the points.
So what is the difference between plt.scatter() vs plt.plot()?

The difference between the two functions is: with pyplot.plot() any property you apply (color, shape, size of points) will be applied across all points whereas in pyplot.scatter() you have more control in each point’s appearance.

That is, in plt.scatter() you can have the color, shape and size of each dot (datapoint) to vary based on another variable. Or even the same variable (y). Whereas, with pyplot.plot(), the properties you set will be applied to all the points in the chart.

අපි දැන් කලින් කිවුවා Scatter plot එකක් කියන්නේ අක්ෂ දෙක දිගේ දත්ත කට්ටල දෙකක ප්‍රස්ථාරයක් කියලා. Scatter plot එකක් විචල්‍ය දෙක අතර සම්බන්ධය පෙන්වීමට භාවිතා කරනවා.

දැන් තවදුරටත් පැහැදිලි කලොත් ඔයාලා දන්නවා Y = mX + C  සරළ ප්‍රස්ථාරයක, X අක්ෂය වැඩි වන විට (හෝ අඩු වන විට) Y අක්ෂය දිගේ අගය වැඩි වන බවක් (හෝ අඩු වන බවක්)  පෙන්වා නම්, එය ධනාත්මක (හෝ සෘණ) රේඛීය සම්බන්ධතාවයක් පෙන්නුම් කරනවා. කෙසේ වෙතත්, කිසිදු පැහැදිලි රටාවක් නැතුව data අහඹු ලෙස බෙදා හරිනු ලැබුවහොත්, එය පරායත්ත සම්බන්ධතාවයක් නොමැතිකම පෙන්නුම් කරන්න පුළුවන්.

````python
import random
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib import animation, rc
from IPython.display import HTML

plt.style.use('fivethirtyeight')

x_vals = []
gold_buy_price = []
gold_sell_price = []

index = count()
initial_price = 1780.43
difference = 0.30

def animate(i):
  x_vals.append(next(index))
  change = round(random.uniform(-0.33, 0.40), 2)
  if len(gold_buy_price)==0:
    gold_buy_price.append(initial_price)
    gold_sell_price.append(initial_price - difference)
  else:
    gold_buy_price.append(gold_buy_price[i-1]+change)
    gold_sell_price.append(gold_sell_price[i-1]+change)
  plt.cla()
  plt.plot(x_vals, gold_buy_price, label='Buy Price')
  plt.plot(x_vals, gold_sell_price, label='Sell Price')
  plt.legend(loc='upper left')
  plt.tight_layout()

ani = animation.FuncAnimation(plt.gcf(), animate, interval=1000)
ani

HTML(ani.to_html5_video())
````

## **8.1  What is a Sub Plot?**
###**Sub Plot එකක් යනු කුමක්ද?** 

When analyzing data you might want to compare multiple plots placed side-by-side. Matplotlib provides a convenient method called subplots to do this.
Subplots mean a group of smaller axes (where each axis is a plot) that can exist together within a single figure. Think of a figure as a canvas that holds multiple plots.

## **8.2 Simple Sub Plot**
## **සරල subplot ප්‍රස්තාරයක් ඇඳීම**

### **8.2.1 Example 1**: Basics of SubPlot
### උදාහරණ 1 :  බහු පදයේ  degree එකේ බලපෑම අනුව අගයේ වෙනස් වීම

````python
import matplotlib.pyplot as plt
import numpy as np


fig,a =  plt.subplots(2,2,figsize=(15,15))

x = np.arange(1,10)

a[0][0].plot(x,x)
a[0][0].set_title('Linear')

a[0][1].plot(x,x*x)
a[0][1].set_title('Quadratic')

a[1][0].plot(x,x*x*x)
a[1][0].set_title('Cubic')

a[1][1].plot(x,np.exp(x))
a[1][1].set_title('Exponential')

plt.show()
````

#### **Example 2**: Answer
#### උදාහරණ 2 : විසඳුම


````python

import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

data = pd.read_csv('/content/sample_data/dev_salary.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, color='#444444',linestyle='--', label='All Devs')
ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')


ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')
ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')


plt.tight_layout()

plt.show()
````

````python
x=np.linspace(1,10,100)

# Plot
plt.plot(x, np.sin(x))
plt.plot(x,np.log(x))

# Decorate
plt.xlabel('x')
plt.title('Sine and Log')
plt.xlim(1,10)
plt.ylim(-1.5 , 3)
plt.show()
````

#### **Exercise 3**: Answer
#### අභ්‍යාසය 3 : විසඳුම

````python
x = np.arange(-10, 10, 0.1)
y1 = x*x
y2 = 1 /(x)

# Plot
fig, ax1 = plt.subplots(figsize=(10,6))
ax1.set_ylabel('x^2')
ax1.plot(x, y1, color='blue')
ax1.set_ylim(0,2)
ax1.grid()

# Second Line on secondary Y axis
ax2 = ax1.twinx()  # join a second axis with the first graph we just made
ax2.set_ylabel('1/x')
ax2.set_ylim(-10,10)
ax2.plot(x, y2,color='red')
plt.show()
````
