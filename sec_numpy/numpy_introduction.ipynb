#Python - NumPy Introduction
Python has been universally used as the go-to language in scientific computing, machine learning applications due to the fact that Python is free, easy to learn, has easy syntax, interpreted nature of execution and a large number of support libraries. However, python is considered a slow language. This means, for a given task it might take more time to do the same workload compared to another language such as C or Java. For applications like AI, deep learning and data processing this leads to a massive disadvantage. 

The following figure shows the speed comparison of python against C language working on the same task. Programs like these are used in DNA sequencing which is very useful in developing vaccines (for example COVID).

As you can see, Python is about 25 times slow against C. So it must be counterintuitive to use such a slow language to do tasks that deal with a huge number of data and take days and weeks to compute. People have identified this bottleneck and developed faster and more efficient libraries to do specific works faster. Using these libraries, we can still use Python (which is a fantastic language for humans to write code) and still produce fast executing programs to quickly do our work.

Among the many libraries python uses to improve its performance, NumPy is one of the most important. Actually, NumPy is the default library to store and process data in computer memory. Also, NumPy is used internally by many other libraries to improve their performance as well.

Numpy mostly deals with arrays. Arrays can be a 1D, 2D or 3D array depending on the application. For a series of data, like time and single variable values the default option is a 1D array. For working with images, a 2D array is the most useful. Also, matrices are a widely used data format in scientific computing stored as 2D arrays. 3D arrays have their own usage we won't go into more details in this course. However, in any case, NumPy provides us with an easy way to store these data and provide functions to process these data very efficiently. There is an example we will go through in the next videos NumPy actually improves the calculation performance and you can verify the advantages of using NumPy arrays yourself.

Like any programming language, Python natively supports an array-like data structure to store your data. You don't need NumPy to use Python lists (arrays). However, when you deal with a large number of data or need a large amount of processing to be done you should consider using the NumPy library and NumPy arrays to save your time. Also, NumPy is a prerequisite to many scientific libraries. They use NumPy internally and the output result of those libraries are provided as a NumPy array. So you have no choice but to use NumPy in those situations.
